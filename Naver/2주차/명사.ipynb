```
post = '''죄책감 덜어진다. 빛의 속도로 삭제했다. 몽쉘같은 초코파이다. 크림과 초코시트가 1:1 비율이다. 부담없이 먹을 수 있다. 속이 더부룩하지 않다. 두툼하고 큼직한 비건 케이크다. 포근하면서 담백하다. 초코의 맛은 첫맛이 강한 편이다. 비건 다크초콜릿이 확 진해지면서 점점 연해진다. 입에서 녹는 몽쉘의 느낌도 있다. 당근케이크도 먹을만 하다.

초코케이크 먹고싶을 때 자주 찾는다. 글루텐프리 전문이다. 매일 당일생산 한다. 크림은 두유로 만든다고 한다. 비건크림이다. 시트사이에 유기농 딸기쨈도 들어가있다. 보들보들해서 돌아서면 생각나는 맛이다.

크게 2가지 메뉴가 있다. LC와 VG로 나뉜다. LC는 저탄수/당뇨식이고 밀가루를 쓰지 않는다. 탄수화물이 적어 당뇨식에 적합하다. VG는 달걀/우유/버터와 같은 동물성 재료를 쓰지 않는다. 초코파이 케이크는 글루텐프리GF&비건VG 음식이다. 참고로 차타고 가면 편한데 걸어가면 계단이 엄청 많다는 점 명심하면 된다. 정말 많다. 겪으면 알게된다.'''
post =  normalize('NFC', re.sub(pattern='[^\w\s]', repl='', string=str(post)))
from khaiii import KhaiiiApi
api = KhaiiiApi()

nng_list=[]
for word in api.analyze(post):
      for morph in word.morphs:
        if morph.tag == 'NNG':
          nng_list.append(morph.lex)
```
