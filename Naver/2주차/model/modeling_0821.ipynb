```
import csv

reviewList = []
with open('preprocessed_data.csv','r') as f: 
  reader = csv.reader(f)
  for line in reader:
    reviewList.append(line[1])
  #확인하기 print(reader)
```

```
from khaiii import KhaiiiApi
api = KhaiiiApi()
```

```
n_tags = ['NNG', 'NNP', 'NNB']#일반, 고유,의존
x_tags = ['XR']

exList = ["정말 맛있고 배도 부르고 특히 마카롱과 커피가 정말 굿입니다.","최고 소름돋는 맛"]
 

def pos_text(texts):
    #for sent in texts:
    n_tagged = ''
    x_tagged = ''
    nList = []
    xList = []
    for word in api.analyze(texts):
      for morph in word.morphs:
        if morph.tag in n_tags:
          n_tagged = morph.lex
          nList.append(n_tagged)
          
          #n_tagged += morph.lex + '/' + morph.tag + ' '
             
        elif morph.tag in x_tags:
          x_tagged = morph.lex
          xList.append(x_tagged)      
    return nList, xList

```
