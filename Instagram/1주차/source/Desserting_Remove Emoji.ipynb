```python
import pandas as pd
import re 
import math

# csv íŒŒì¼ ì½ì–´ì˜¤ê¸°
raw_data = pd.read_csv('Instagram Crawling.csv', encoding='utf-8').drop('Unnamed: 0', axis=1)

raw_data.head()
```

```python
# emojië¥¼ ì§€ìš°ëŠ” í•¨ìˆ˜

def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    
    return emoji_pattern.sub(r'', string)
```

```python
# Reviewì™€ Hashtagì— ì ìš©

for i in range(len(raw_data)):
    raw_data['Review'][i] = remove_emoji(str(raw_data['Review'][i]))
    raw_data['Hashtag'][i] = remove_emoji(str(raw_data['Hashtag'][i]))
```

```python
# Hashtagì˜ #ì„ ì§€ìš°ëŠ” ì½”ë“œ

for i in range(len(raw_data)):
    if raw_data['Hashtag'][i] == "nan": # Hashtagê°€ ì•„ë¬´ê²ƒë„ ì—†ì„ ê²½ìš°
        raw_data['Hashtag'][i] = None   # Noneìœ¼ë¡œ ë³€ê²½
    else:
        raw_data['Hashtag'][i] = raw_data.Hashtag[i][1:].replace("#",",")
```
