```
from khaiii import KhaiiiApi

api = KhaiiiApi()
morp_dic = {}
need_morp = ['VV','VA','VX']
#morph.lex, morph.tag

for i in need_morp:
  morp_dic[i]=[]

for sentence in data.Review[0]:
  try:
    for word in api.analyze(sentence):
      for morph in word.morphs:
        tag = morph.tag
        if tag in need_morp:
          morp_dic[tag].append(morph.lex)
  except:
    continue

from collections import Counter
for i in need_morp:
  morp_dic[i] = Counter(morp_dic[i])
```

```
api = KhaiiiApi()
morp_dic = {}
need_morp = ['VV','VA','VX']
#morph.lex, morph.tag

for i in need_morp:
  morp_dic[i]=[]

for i in range(10) :

  for sentence in data.Review[i]:
    try:
      for word in api.analyze(sentence):
        for morph in word.morphs:
          tag = morph.tag
          if tag in need_morp:
            morp_dic[tag].append(morph.lex)

      vv = []
      for v in set(morp_dic['VV']) :
        vv.append(v)
      va = []
      for a in set(morp_dic['VA']) :
        va.append(a)
      vx = []
      for x in set(morp_dic['VX']) :
        vx.append(x)

      for k in range(len(vv)) :
        vv[k] = vv[k] in list(dic.VV)
      for t in range(len(va)) :
        va[t] = vv[t] in list(dic.VA)
      for s in range(len(vx)) :
        vx[s] = vx[s] in list(dic.VX)

      if sum(vv)+sum(va)+sum(vx) >= 2 :
        data.Sentiment[i] =1
    except:
      continue
```
